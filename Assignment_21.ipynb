```python
with open('today.txt', 'w') as f:
    f.write(str(datetime.today().date()))
```

<b>6. Use multiprocessing to create three separate processes. Make each one wait a random number of
seconds between one and five, print the current time, and then exit. 

```python
def current_time():
    time.sleep(np.random.randint(1,5,1)[0])
    
p1 = mp.Process(target=current_time)
p2 = mp.Process(target=current_time)
print(str(datetime.now().time()))
p1.start()
p2.start()
p1.join()
p2.join()
p1.kill()
p2.kill()
print(str(datetime.now().time()))
```
